```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/PlayingNumbers/ds_salary_proj/master/glassdoor_jobs.csv')
```

```python
df
```

# Data Cleaning

* Salary parsing
* Company name text only
* State field
* Age of company
* Parsing of job description (phyton, etc)

### Salary parsing

```python
# Removing meaningless data

df = df[df['Salary Estimate'] != '-1' ]
df
```

```python
# Making new column for unique Salary Estimate

df['Hourly'] = df['Salary Estimate'].apply(lambda x : 1 if 'per hour' in x.lower() else 0)
df['Employer Provided'] = df['Salary Estimate'].apply(lambda x : 1 if 'employer provided salary' in x.lower () else 0)
```

```python
# Making salary estimate field more meaningful

salary = df['Salary Estimate'].apply(lambda x: x.split('(') [0])
salary
```

```python
minus_kd = salary.apply(lambda x : x.replace ('K', '').replace ('$', ''))
minus_kd
```

```python
minus_hr = minus_kd.apply(lambda x : x.lower().replace('per hour', '').replace('employer provided salary:',''))
```

```python
# Making Min and Max salary column to get the average salary

df['Min Salary'] = minus_hr.apply(lambda x : int(x.split('-')[0]))
df['Max Salary'] = minus_hr.apply(lambda x : int(x.split('-')[1]))
df
```

```python
df['Avg Salary'] = (df['Min Salary']+df['Max Salary']) / 2
df
```

### Company name text only

```python
# Removing rating from Company Name

df['Company Txt'] = df.apply(lambda x : x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3], axis = 1)
df
```

### State field


```python
# Splitting states from location

df['Job State'] = df['Location'].apply(lambda x : x.split(',')[1])
df
```

```python
# Checking the avaibility of jobs based on states

df['Job State'].value_counts()
```

```python
# Making new column to find if job location and company headquarters is the same or not
df['Same State'] = df.apply(lambda x : 1 if x.Location == x.Headquarters else 0, axis =1)
df
```

### Age of company

```python
# Making new column for company age

df['Company Age'] = df['Founded'].apply(lambda x : x if x < 0 else 2021 - x)
df
```

### Parsing of job description

```python
# Parsing job description for data science tools that I have learned (Python, SQL, Tableu, Excel)

df['Python'] = df['Job Description'].apply(lambda x : 1 if 'python' in x.lower() else 0)
df['Python'].value_counts()
```

```python
df['SQL'] = df['Job Description'].apply(lambda x : 1 if 'sql' in x.lower() else 0)
df['SQL'].value_counts()
```

```python
df['Tableu'] = df['Job Description'].apply(lambda x : 1 if 'tableu' in x.lower() else 0)
df['Tableu'].value_counts()
```

```python
df['Excel'] = df['Job Description'].apply(lambda x : 1 if 'excel' in x.lower() else 0)
df['Excel'].value_counts()
```
